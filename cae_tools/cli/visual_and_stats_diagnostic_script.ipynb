```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Reading the data
ds_train_toy = xr.open_dataset("C:\\GitHub\Downscaling\\cae_tools_v2_16Apr2024\\cae_tools\\cae_tools\\train_scores.nc")
ds_test_toy = xr.open_dataset("C:\\GitHub\Downscaling\\cae_tools_v2_16Apr2024\\cae_tools\\cae_tools\\test_scores.nc")

print(ds_train_toy.box.shape)

print(ds_test_toy.box.shape)



print(ds_train_toy)

print(ds_test_toy)




```

```python
num_box = ds_train_toy.hires_estimate.shape[0]
mse = np.zeros(num_box)
vmin = np.zeros(num_box)
vmax = np.zeros(num_box)
rmse_values = []

for n in np.linspace(0,num_box-1,40,dtype=int): #range
    plt.figure(figsize=(12,4))  # Set a larger figure size
    # vmin = min(ds_train_toy.hires_estimate[n, 0, :, :].min(), ds_train_toy.hires[n, 0, :, :].min())
    # vmax = max(ds_train_toy.hires_estimate[n, 0, :, :].max(), ds_train_toy.hires[n, 0, :, :].max())
    # vmin = 288
    # vmax = 293
    vmin[n] = ds_train_toy.ST_slices[n, 0, :, :].min()
    vmax[n] = ds_train_toy.ST_slices[n, 0, :, :].max()    

    # Plot the estimated high resolution
    plt.subplot(1, 3, 1)  # 1 row, 2 columns, 1st subplot
    plt.pcolormesh(ds_train_toy.hires_estimate[n, 0, :, :], cmap='jet', vmin=vmin[n], vmax=vmax[n])          # , vmin=vmin[n], vmax=vmax[n])
    plt.colorbar()
    plt.title(f"Estimated High Res Box {n}")

    # Plot the actual high resolution
    plt.subplot(1, 3, 2)  # 1 row, 2 columns, 2nd subplot
    plt.pcolormesh(ds_train_toy.ST_slices[n, 0, :, :], cmap='jet')                   # , vmin=vmin[n], vmax=vmax[n])
    plt.colorbar()
    plt.title(f"Actual High Res Box {n}")

    # Plot the land cover
    plt.subplot(1, 3, 3)  # 1 row, 2 columns, 2nd subplot
    plt.pcolormesh(ds_train_toy.land_cover[n, 0, :, :], cmap='jet')                   # , vmin=vmin[n], vmax=vmax[n])
    plt.colorbar()
    plt.title(f"Land Cover Box {n}")    

    plt.show()

    plt.figure(figsize=(12,4))  # Set a larger figure size
    plt.subplot(1, 3, 1)  # 1 row, 2 columns, 1st subplot
    plt.pcolormesh(ds_train_toy.tasmax[n, 0, :, :], cmap='jet')          # , vmin=vmin[n], vmax=vmax[n])
    plt.colorbar()
    plt.title(f"Tasmax Box {n}")

    # Plot the actual high resolution
    plt.subplot(1, 3, 2)  # 1 row, 2 columns, 2nd subplot
    plt.pcolormesh(ds_train_toy.tasmin[n, 0, :, :], cmap='jet')                   # , vmin=vmin[n], vmax=vmax[n])
    plt.colorbar()
    plt.title(f"Tasmin Box {n}")

    # Plot the land cover
    plt.subplot(1, 3, 3)  # 1 row, 2 columns, 2nd subplot
    plt.pcolormesh(ds_train_toy.rainfall[n, 0, :, :], cmap='jet')                   # , vmin=vmin[n], vmax=vmax[n])
    plt.colorbar()
    plt.title(f"Rainfall Box {n}")  

    plt.show() 

    # # Compute MSE with NumPy
    # mse[n] = np.mean((ds_train_toy.hires_estimate[n, 0, :, :].values - ds_train_toy.ST_slices[n, 0, :, :].values) ** 2)
    # print(f"MSE for box {n}: {mse[n]}")  
    print(f"box time is {ds_train_toy.box_time[n].values}")   
    # print(f"box SW corner indeces {ds_train_toy.SW_corner_y[n].values,ds_test_toy.SW_corner_x[n].values}")

    # Compute RMSE:
    difference = ds_train_toy.hires_estimate[n, 0, :, :] - ds_train_toy.ST_slices[n, 0, :, :]
    mse = np.mean(difference ** 2)
    rmse = np.sqrt(mse)
    rmse_values.append(rmse)

plt.plot(rmse_values)
plt.xlabel("Box number")
plt.ylabel("RMSE")
plt.title("RMSE for each box")
plt.show()

    
```

```python
import numpy as np
from skimage.metrics import structural_similarity as ssim

# Assuming ds_train_toy.hires_estimate and ds_train_toy.ST_slices are xarray DataArrays
num_boxes = ds_train_toy.hires_estimate.shape[0]

ssim_values = []
correlation_values = []

for i in range(200):
    # Extract the image arrays and convert them to NumPy arrays
    image_true = ds_train_toy.ST_slices[i, 0, :, :].values
    image_test = ds_train_toy.hires_estimate[i, 0, :, :].values

    # Calculate SSIM
    current_ssim = ssim(image_true, image_test, data_range=image_test.max() - image_test.min())
    ssim_values.append(current_ssim)

    # Flatten the images to compute correlation
    image_true_flat = image_true.flatten()
    image_test_flat = image_test.flatten()
    correlation = np.corrcoef(image_true_flat, image_test_flat)[0, 1]
    correlation_values.append(correlation)

    # print(f"Box ID {i} - SSIM: {current_ssim}, Correlation Coefficient: {correlation}")

# Print mean values if needed
mean_ssim = np.mean(ssim_values)
mean_correlation = np.mean(correlation_values)
print(f"Mean SSIM: {mean_ssim}, Mean Correlation: {mean_correlation}")

```
